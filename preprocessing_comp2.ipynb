Las suma de las columnas ``mcomisiones``, ``mactivos_margen`` y ``mpasivos_margen`` parecen corresponderse con ``mrentabilidad``, que corresponde a la rentabilidad total del mes del banco por el cliente.

Vamos a proyectar la ganancia anual para aquellos clientes que llevan menos de 1 año en el banco

```python
# Simplemente proyectamos la rentabilidad mensual

data["mrentabilidad_annual"] = data[
    ["mrentabilidad_annual", "cliente_antiguedad", "mrentabilidad"]
].apply(
    lambda row: row["mrentabilidad_annual"]
    if row["cliente_antiguedad"] >= 12
    else row["mrentabilidad_annual"]
    + (12 - row["cliente_antiguedad"]) * row["mrentabilidad"],
    axis=1,
)
```

```python
# Corregimos status
import numpy as np

def fix_credit_status(status):
    if np.isnan(status):
        return status
    if status == 0:
        return status
    if status == 6:
        return 1
    if status == 7:
        return 2
    if status == 9:
        return 3
    raise NotImplementedError(f"Hay un status {status} diferente a la documentacion")

data["Master_status"] = data["Master_status"].apply(fix_credit_status)
data["Visa_status"] = data["Visa_status"].apply(fix_credit_status)
```

```python
def delinquency(delinquency_status_visa, deliquency_status_master):
    if np.isnan(delinquency_status_visa):
        return deliquency_status_master
    if np.isnan(deliquency_status_master):
        return delinquency_status_visa
    return int(delinquency_status_visa or deliquency_status_master)
```

```python
def check_negative(data_col):
    try:
        return (data_col < 0).any()
    except TypeError:
        return False

def split_in_negative_and_positive(x):
    if x <= 0:
        return abs(x), 0
    return 0, x

def fix_negative_and_positive(data: pd.DataFrame):
    cols_with_negatives = [col for col in data.columns if check_negative(data[col])]
    for col in cols_with_negatives:
        data[f"{col}_negative"], data[f"{col}_postive"] = zip(*data[col].map(split_in_negative_and_positive))
        data = data.drop(columns = col)
    return data
```

```python
# Hacemos un split en positivo/negativo
data = fix_negative_and_positive(data)
```

```python
def _create_ratios_monto_cantidad(monto, cantidad):
    if monto == 0 or cantidad == 0:
        return 0
    return monto/cantidad

def create_ratios_monto_cantidad(data: pd.DataFrame):
    montos_columns = [c for c in data.columns if c.startswith("m") and "negative" not in c and "positive" not in c]
    for monto_col in montos_columns:
        try:
            data[f"ratio_{monto_col[1:]}"] = data[[monto_col, "c"+monto_col[1:]]].apply(
                                lambda row: _create_ratios_monto_cantidad(row[monto_col], row["c"+monto_col[1:]]), axis = 1)
        except KeyError:
            continue
    return data
    
```

```python
data = create_ratios_monto_cantidad(data)
```

```python
def corregir_inflacion(data):
    inflacion_marzo_mayo = (1.041)*(1.033)
    monto_columns = [c for c in data.columns if c.startswith("m") and "negative" not in c and "positive" not in c]
    for col in monto_columns:
        data.loc[data["foto_mes"] == 202105, col] = data.loc[data["foto_mes"] == 202105, col]/inflacion_marzo_mayo
    return data
    
```
